```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVR


# Model ML
svr  = SVR(C=1000, gamma=0.05, kernel='rbf')

df = pd.read_csv('Salary_Data.csv')

# Memisahkan data
x = []
for i in df['YearsExperience']:
    x.append([i])
y = df['Salary']

# Train
svr.fit(x, y)

# View
plt.scatter(x, y)
plt.plot(x, svr.predict(x))
```
